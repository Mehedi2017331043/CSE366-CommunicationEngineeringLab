#<center> **<h2>Signals and Plotting the signal.**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
t = np.linspace(0, 2.0, 100)
def create_sinusoidal(A, f, t, phase = 0):
  x = A * np.sin(2.0 * np.pi * f * t + phase)
  return x
```

The working equation is:<br>
<center>$$x(t) = A\sin(2\pi f t + phase)$$

```python
x = create_sinusoidal(1,1,t)
sns.set()

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(18, 18))
fig.suptitle('\nDifferent subplots of a sinusoidal signal', fontsize=25)

ax[0, 0].plot(t,x,'g')
ax[0, 0].set_title('Continuous plot', fontsize = 25)

ax[0, 1].stem(t,x,'g')
ax[0, 1].set_title('Stem plot', fontsize = 25)

ax[1,0].scatter(t,x)
ax[1,0].set_title('Scatter plot', fontsize = 25)
ax[-1,-1].axis('off')
for a in ax.flat:
    a.set(xlabel='time(t)', ylabel='x(t)')

plt.show()
```



> **Composite signal** is a combination of simple sine waves with different frequencies, amplitudes, and phases.A single-frequency sine wave is not useful in data communications;we need to send a composite signal, a signal made of many simple sine waves.




```python
x1 = create_sinusoidal(1,3,t)
x2 = create_sinusoidal(2,5,t)
x3 = create_sinusoidal(3,7,t)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(18, 18))
fig.suptitle('3 Sinusoidal signal and a composite signal', fontsize=25)

ax[0, 0].plot(t,x1, 'g')
ax[0, 0].set_title('x1(t)', fontsize = 25)

ax[0, 1].plot(t,x2, 'g')
ax[0, 1].set_title('x2(t)', fontsize = 25)

ax[1,0].plot(t,x3, 'g')
ax[1,0].set_title('x3(t)', fontsize = 25)

ax[1,1].plot(t, x1 + x2 + x3,'g')
ax[1,1].set_title('y(t) = x1(t) + x2(t) + x3(t)', fontsize = 25)
plt.show()
```

The working equation is:<br>
<center>$$A\sin(2\pi f t + phase)=x(t)  \to x(n)$$

```python
fs = 10
step = 1.0/fs
n = np.arange(0, 2.1, step)
y = create_sinusoidal(1,5/fs,n)
plt.title('fs = 10', fontsize = 25)
plt.stem (n,y,'g')
plt.show()
```

#<center>**Basic Properties of a Signal.**

```python
def Shifting(y,k,n):
  ret=np.zeros(abs(k))

  if(k>=0):
    ret=np.concatenate((ret,y[0:len(n)-k]))
  else:
    ret=np.concatenate((y[abs(k):len(n)],ret))
  return ret
```

```python
k=5
plt.stem(n,Shifting(y,k,n),'g')
plt.title('Delay', fontsize = 15)
plt.show()

k=-5
plt.stem(n,Shifting(y,k,n),'g')
plt.title('Advance', fontsize = 15)
plt.show()
```



> **Folding or Reflection**

Folding or Reflection of the signal about the time origin n = 0 is the time base replacement of the independent variable n by n.



```python
plt.stem(n,y,'g')
plt.title("Orginal Signal")
plt.show()

plt.stem(-n,y,'g')
plt.title("Folded Signal")
plt.show()
```
