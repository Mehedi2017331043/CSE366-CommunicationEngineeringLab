**<h3>Amplitude Modulation(AM):**



> **Amplitude modulation (AM)** is a modulation technique used in electronic communication, most commonly for transmitting messages with a radio wave. In amplitude modulation, the amplitude (signal strength) of the carrier wave is varied in proportion to that of the message signal, such as an audio signal.



```python
def plot_signal(t, x):

  plt.figure(figsize=(20, 6))
  plt.plot(t, x,'b')
  plt.show()
```



> **Step 1: Creating Carrier Signal**



The working equation is:<br>
<center>$$c(t)=Asin(2Ï€fct)$$


```python
import numpy as np
import matplotlib.pyplot as plt
import math
A=2; 
fc=0.4; 
t=np.arange(0, 50, 0.1)
ct=A*np.sin(2*np.pi*fc*t); 
plot_signal(t, ct)
```



> **Step 2: Creating message Signal**




The working equation is:<br>
<center>$$m(t) = Mcos(2{\pi}f_{m}t+{\phi})$$$$m(t) = Amcos(2{\pi}f_{m}t+{\phi})$$


```python
m = 1.0
M = A * m
fm=.05
phase=0
mt=M*np.cos((2*np.pi*fm*t)+phase)
plot_signal(t, mt)
```





> **Step 3: Computing Amplitude modulation**





The working equation is:<br>
<center>$$y(t) = [1 + mcos(2{\pi}f_{m}t+{\phi})]*Asin(2{\pi}f_{c}t)$$

```python
yt=(1+m*np.cos((2*np.pi*fm*t) + phase))*A*np.sin(2*np.pi*fc*t);
plot_signal(t, yt)
```

# **<h2>Frequency Modulation(FM)**#



> Frequency Modulation (FM) is the encoding of information in a carrier wave by changing the instantaneous frequency of the wave.






> **Step 1: Creating Carrier Signal**





The working equation is:<br>
<center>$$x_{c}(t)=A_{c}cos(2{\pi}f_{c}t)$$

```python
Ac=2; 
fc=30; 
t=np.arange(0, 1.0, 0.0001)
xc=Ac*np.cos(2*np.pi*fc*t); 
plot_signal(t, xc)
```



> **Step 2: Creating modulating signal**



```python
m = 1.
fm=5
xm=np.sin((2*np.pi*fm*t))
plot_signal(t, xm)
```



> **Step 3: Computing Frequency modulation**



```python
yt = Ac*np.cos(2. * np.pi * (fc * t + m * np.sin(2*np.pi*fm*t)))
plot_signal(t, yt)
```

# **Phase Modulation(PM)**



> Phase modulation (PM) is a modulation pattern for conditioning communication signals for transmission.





> **Step 1: Creating Carrier Signal**



The working equation is:<br>
<center>$$c(t)=A_{c}sin(2{\pi}f_{c}t + {\phi_{c}})$$


```python
Ac=2; 
fc=4; 
t=np.arange(0, 4.0, 0.0001)
phase = 0
ct=Ac*np.sin(2*np.pi*fc*t + phase); 
plot_signal(t, ct)
```



> **Step 2: Creating message Signal**


The working equation is:<br>
<center>$$m(t) = A_{m}sin(2{\pi}f_{m}t)$$

```python
k = 1.0
fm=3
Am = 2
mt=Am*np.sin(2 * np.pi*fm*t)
plot_signal(t, mt)
```



> **Step 3: Computing Phase modulation**



The working equation is:<br>
<center>$$y(t) = A_{c}sin(2{\pi}f_{c}t+ km(t) +{\phi_{c}})   $$

```python
yt=Ac*np.sin((2*np.pi*fc*t) + (k * mt + phase))
plot_signal(t, yt)
```
